```python
V0_CASE_ID_COL_NAME = "case_id"
V0_SCENARIO_ID_1_COL_NAME = "scenario_id_1"
V0_SCENARIO_ID_2_COL_NAME = "scenario_id_2"
V0_SCENARIO_ID_3_COL_NAME = "scenario_id_3"
```

```python
df_v0_cols = df_v0.columns
for col_i, col_name in enumerate(df_v0_cols):
  print(col_i, col_name)
```

```python
def remap_cols(col_names: [str], new_col_idxs: [int]):
  ans_col_names = []
  for new_col_idx in new_col_idxs:
    ans_col_names.append(col_names[new_col_idx])
  return ans_col_names
```

```python
# 0 condition_id X
# 1 scenario_id_1
# 2 scenario_id_2
# 3 scenario_id_3
# 4 case_id
# 5 player_limit X
# 6 ai_player_limit X
# 7 round_limit X
# 8 game_limit X
# 9 survey_delay X
# 10 scenario_some_id_1
# 11 information_card_id_1
# 12 comp_information_card_ID_1
# 13 game_end_survey_ID_1
# 14 feedback_ID_1
# 15 scenario_some_id_2
# 16 information_card_id_2
# 17 comp_information_card_ID_2
# 18 game_end_survey_id_2
# 19 feedback_id_2
# 20 scenario_some_id_3
# 21 information_card_id_3
# 22 comp_information_card_id_3
# 23 game_end_survey_ID_3
# 24 feedback_id_3
# 25 game_scenario_description_1
# 26 game_scenario_description_2
# 27 game_scenario_description_3
# 28 ai_p0_types
# 29 ai_p1_types
# 30 midgame_id_1
# 31 midgame_id_2
# 32 midgame_id_3
# 33 feedback_type
# 34 ai_player_types_id
df_v1_cols = remap_cols(df_v0_cols, [4, 1, 11, 12, 13, 14, 2, 16, 17, 18, 19, 3, 21, 22, 23, 24])
df_v1_cols
```

```python
df_v1 = df_v0[df_v1_cols]
df_v1
```

```python
def transform_scenario_v0_id_str_fn(name_str:str):
  ans_list = [0,0,0]
  parsing_num_flag = False
  num_i = 0
  for c_char in name_str:
    if c_char.isdigit():
      if num_i > 2:
        print(name_str)
      ans_list[num_i] = ans_list[num_i] * 10 + int(c_char)
      parsing_num_flag = True
    else:
      if parsing_num_flag == True:
        parsing_num_flag = False
        num_i += 1

  return (ans_list[0], ans_list[1], ans_list[2])

assert(transform_scenario_v0_id_str_fn("S1T1C1") == (1,1,1))
assert(transform_scenario_v0_id_str_fn("S1T2C2") == (1,2,2))
assert(transform_scenario_v0_id_str_fn("S1T2") == (1,2,0))
```

```python
def transform_case_v0_id_str_fn(case_v0_id_str: str):
  CHAR_LOWER_A_U8 = ord('a')
  if len(case_v0_id_str) < 5:
    return (int(case_v0_id_str[1]), int(case_v0_id_str[3]), 0)
  else:
    return (int(case_v0_id_str[1]), int(case_v0_id_str[3]), int(ord(case_v0_id_str[4]) - CHAR_LOWER_A_U8))

assert(transform_case_v0_id_str_fn("C1D1") == (1,1,0))
assert(transform_case_v0_id_str_fn("C1D1a") == (1,1,0))
assert(transform_case_v0_id_str_fn("C2D2b") == (2,2,1))
```

```python
for row_idx in range(0,len(df_v0)):
  row = df_v0.iloc[row_idx]
  condition_info = transform_case_v0_id_str_fn(row[V0_CASE_ID_COL_NAME])
  scenario_1_info = transform_scenario_v0_id_str_fn(row[V0_SCENARIO_ID_1_COL_NAME])
  scenario_2_info = transform_scenario_v0_id_str_fn(row[V0_SCENARIO_ID_2_COL_NAME])
  scenario_3_info = transform_scenario_v0_id_str_fn(row[V0_SCENARIO_ID_3_COL_NAME])
  print(f"[{row_idx + 1},{condition_info[1]},{condition_info[2]},{scenario_1_info[0]},{scenario_1_info[1]},{scenario_2_info[0]},{scenario_2_info[1]},{scenario_3_info[0]},{scenario_3_info[1]}],")

```
