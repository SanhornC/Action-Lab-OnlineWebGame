## Cleaning Column `case_id`

It seems that all the `case_id` start with "C1".

It seems there are no cases with `case_id` "C1D1" starts with anything other than "a".

```python
def transform_case_v0_id_str_fn(case_v0_id_str: str):
  CHAR_LOWER_A_U8 = ord('a')
  return (int(case_v0_id_str[1]), int(case_v0_id_str[3]), int(ord(case_v0_id_str[4]) - CHAR_LOWER_A_U8))

assert(transform_case_v0_id_str_fn("C1D1a") == (1,1,0))
assert(transform_case_v0_id_str_fn("C2D2b") == (2,2,1))
```

## Parse Things like "S1T1C1"

```python
def transform_scenario_v0_id_str_fn(name_str:str):
  ans_list = [0,0,0]
  parsing_num_flag = False
  num_i = 0
  for c_char in name_str:
    if c_char.isdigit():
      if num_i > 2:
        print(name_str)
      ans_list[num_i] = ans_list[num_i] * 10 + int(c_char)
      parsing_num_flag = True
    else:
      if parsing_num_flag == True:
        parsing_num_flag = False
        num_i += 1

  return (ans_list[0], ans_list[1], ans_list[2])

assert(transform_scenario_v0_id_str_fn("S1T1C1") == (1,1,1))
assert(transform_scenario_v0_id_str_fn("S1T2C2") == (1,2,2))
assert(transform_scenario_v0_id_str_fn("S1T2") == (1,2,0))
```

## Parse Card Indices String

```python
def transform_card_id_list_str_to_idx_fn(card_id_list_string: str):
  card_id_list = card_id_list_string.split(',')
  card_idx_list = []
  for card_id_str in card_id_list:
    card_idx_list.append(int(transform_scenario_v0_id_str_fn(card_id_str)[2]) - 1) # To Zero Based Index
  card_idx_list.sort()
  return card_idx_list

assert(transform_card_id_list_str_to_idx_fn("S1T1C1,S1T1C2") == [0,1])
assert(transform_card_id_list_str_to_idx_fn("S1T2C1,S1T2C3") == [0,2])
```

## Parse Game Num Index String 

```python
def transform_in_game_num_idx_str_to_idx_fn(game_num_idx_str: str):
  return int(game_num_idx_str[-1]) - 1

assert(transform_in_game_num_idx_str_to_idx_fn("C36_S1") == 0)
assert(transform_in_game_num_idx_str_to_idx_fn("C36_S2") == 1)
```

```python
def transform_condition_id_str_to_idx_fn(condition_id_str: str):
  ans_condition_i = 0
  met_number_c_flag = False
  for c in condition_id_str:
    if c.isdigit():
      ans_condition_i = ans_condition_i * 10 + int(c)
      met_number_c_flag = True
    elif met_number_c_flag == True:
      break
  return ans_condition_i

assert(transform_condition_id_str_to_idx_fn("C36_S1") == 36)
assert(transform_condition_id_str_to_idx_fn("C35_S2") == 35)
```

```python
# case d, case variant, scenario id, scenario type, player_0 card indices, player_1 card indices
# info_map = {}
# duplicate_count = 0

info_list = []

for row_idx in range(0,len(df_v0)):
  row = df_v0.iloc[row_idx]
  case_info_tuple = transform_case_v0_id_str_fn(row[V0_COL_CASE_ID])
  scenario_info_tupe = transform_scenario_v0_id_str_fn(row[V0_COL_SCENARIO_ID])
  condition_idx = transform_condition_id_str_to_idx_fn(row[V0_COL_GAME_NUM])
  game_num_idx = transform_in_game_num_idx_str_to_idx_fn(row[V0_COL_GAME_NUM])
  player_0_card_idx_list = transform_card_id_list_str_to_idx_fn(row[V0_COL_PLAYER_0_CARDS_ID])
  player_1_card_idx_list = transform_card_id_list_str_to_idx_fn(row[V0_COL_PLAYER_1_CARDS_ID])
  assert(len(player_0_card_idx_list) == len(player_1_card_idx_list))
  info_list.append([condition_idx, game_num_idx, case_info_tuple[1],case_info_tuple[2], scenario_info_tupe[0], scenario_info_tupe[1], player_0_card_idx_list, player_1_card_idx_list])
  # player_card_id_info = (case_info_tuple[1],case_info_tuple[2], scenario_info_tupe[0], scenario_info_tupe[1], game_num_idx)
  # if not player_card_id_info in info_map.keys():
  #   info_map[player_card_id_info] = (player_0_card_idx_list,player_1_card_idx_list)
  # else:
  #   assert(info_map[player_card_id_info] == (player_0_card_idx_list, player_1_card_idx_list))
  #   duplicate_count += 1

# print(f"Duplicate: {duplicate_count} / {len(df_v0)}")
```
